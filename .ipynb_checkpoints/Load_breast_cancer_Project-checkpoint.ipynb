```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
cancer_df = pd.DataFrame(cancer["data"] , columns=cancer["feature_names"])
cancer_df
```

```python
cancer_df["target"] = cancer["target"]
cancer_df
```

```python
def evaluate_preds(y_true , y_pred):
    """

    """
    acc = accuracy_score(y_true , y_pred)
    recall = recall_score(y_true , y_pred)
    precision = precision_score(y_true , y_pred)
    f1 = f1_score(y_true , y_pred)

    metric_dict = {"Acc":round(acc,2),
                   "recall":round(recall,2),
                   "precision":round(precision,2),
                   "f1":round(f1,2)}

    print(f"The accuracy score is:{acc*100:.2f}%")
    print(f"The recall score is :{recall:.2f}")
    print(f"The precision score is :{precision:.2f}")
    print(f"The f1 score is :{f1:.2f}")

    return metric_dict
```

```python
#Hyperparameters by hand
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , recall_score , precision_score ,f1_score
from sklearn.model_selection import train_test_split

np.random.seed(42)

cancer_df_shuffled = cancer_df.sample(frac=1)

x = cancer_df_shuffled.drop("target" ,axis=1)
y = cancer_df_shuffled["target"]

train_split = round(0.7*len(cancer_df_shuffled))
valid_split = round(train_split + 0.15 *len(cancer_df_shuffled))
x_train , y_train = x[:train_split],y[:train_split]
x_valid , y_valid = x[train_split:valid_split],y[train_split:valid_split]
x_test , y_test = x[valid_split:],y[valid_split:]

clf = RandomForestClassifier()
clf.fit(x_train , y_train)

y_preds = clf.predict(x_valid)

baseline_metrics = evaluate_preds(y_valid , y_preds)
baseline_metrics
```

```python
clf_2 = RandomForestClassifier(n_estimators=100)
clf_2.fit(x_train , y_train)

y_preds_2 = clf_2.predict(x_valid)
clf_2_metrics = evaluate_preds(y_valid , y_preds_2)
```

```python
#Using RadnomizedSearchCV hyperparmeters
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV , train_test_split

np.random.seed(42)

grid = {"n_estimators":[10,100,200,500,1000,1200],
        "max_depth":[None , 5,10,20,30],
        "max_features":["log2","sqrt"],
        "min_samples_split":[2,4,6],
        "min_samples_leaf":[1,2,4]}

cancer_df_shuffled = cancer_df.sample(frac=1)

x = cancer_df_shuffled.drop("target" , axis=1)
y = cancer_df_shuffled["target"]

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)

clf = RandomForestClassifier(n_jobs=1)

rs_clf = RandomizedSearchCV(clf ,
                            param_distributions=grid,
                            cv=5,
                            n_iter=10,
                            verbose=2)
rs_clf.fit(x_train , y_train)
```

```python
rs_y_preds = rs_clf.predict(x_test)
rs_metrics = evaluate_preds(y_test , rs_y_preds)
```

```python
#Using GridSearchCV Hyperparameters
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV , train_test_split

np.random.seed(42)

grid_2 = {'n_estimators': [100,200,500],
         'min_samples_split': [6],
         'min_samples_leaf': [1,2],
         'max_features': ['log2','sqrt'],
         'max_depth': [None]}

cancer_df_shuffled = cancer_df.sample(frac=1)

x = cancer_df_shuffled.drop("target" , axis=1)
y = cancer_df_shuffled["target"]

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)

clf = RandomForestClassifier(n_jobs=1)

gs_clf = GridSearchCV(clf ,
                      param_grid=grid_2,
                      cv=5,
                      verbose=2)
gs_clf.fit(x_train , y_train)
```

```python
gs_y_preds = gs_clf.predict(x_test)
gs_metrics = evaluate_preds(y_test , gs_y_preds)
```

```python
compare_metrics=pd.DataFrame({"Baseline_metrics":baseline_metrics,
                                "clf_2_metrics":clf_2_metrics,
                                "rs_metrics":rs_metrics,
                                "gs_metrics":gs_metrics})
compare_metrics.plot.bar(figsize=(10,6))
plt.show()
```
